```python
x = np.arange(5)
y = np.arange(10,5,-1)
plt.scatter(x,y)
plt.show()

x = np.linspace(0,10,30)
y = x**2
plt.scatter(x,y,c=y,cmap = "plasma")
plt.show()

color = np.linspace(20,30,30)
plt.scatter(x,y,c=color,s=50)
plt.axis("square")
plt.title("WOW FAJNE KROPY")
plt.legend(y)
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-np.pi,np.pi,20),np.linspace(-np.pi/4,np.pi/4,10))
z = np.sin(x) * np.cos(y)
plt.contourf(x,y,z,50)
plt.show()

z = np.sin(x*2) + np.cos(y) + np.sin(x*y)
plt.contourf(x,y,z,100)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
surf = ax.plot_surface(x,y,z,cmap = cm.coolwarm, antialiased = False)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title("WOW 3D")
plt.show()
```

```python
x,y = np.meshgrid(np.arange(0,100),np.arange(200,300))
z = x**y
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.plot_surface(x,y,z,cmap=cm.plasma)
ax.plot_surface(x,z,y,cmap=cm.coolwarm)
```

```python
x = np.arange(-4,6,2)
x = np.linspace(-4,6,100)
for k in range(1,6):
    l = k/10
    y = [np.pi**(i*l) for i in x]
    plt.plot(x,y,label="$pi^{%sx}$" % l,linewidth=2)
plt.legend()
plt.grid()
plt.show()

x,y = np.meshgrid(np.linspace(0,100,1000),np.linspace(0,100,100))
z = x*y**2
plt.contourf(x,y,z,800,cmap="cividis")
plt.axis("square")
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.plot_surface(x,y,z,cmap=cm.magma)
plt.show()
```
