```python
def julia(z,c):
    for i in range(100):
        z=z*z+c
        if abs(z) > 2 : return i
    return 0

c=-0.8+ 0.156j
x = []
y = []
color=[]
d = 600

for i in range(d):
    for j in range(d):
        xi=-1.5+3/d*i
        yj=-1.5+3/d*j
        color.append(julia(xi+yj*1j,c))
        x.append(xi)
        y.append(yj)


plt.figure(figsize=(10,10))
plt.scatter(x,y,s=0.1,c=color)
plt.axis('square')
plt.show()
```

```python
def julia(z,c): #z jest tablicą
    for i in range(200):
        z=z*z+c
        if abs(z)>2: return i
    return 0
c=-0.1+ 0.65j
d = 1000
X,Y = np.meshgrid(np.linspace(-1.5,1.5,d),np.linspace(-1.5,1.5,d))

color= np.frompyfunc(julia,2,1)(X+Y*1j,c).astype(np.float64)

plt.figure(figsize=(10,10))
plt.contourf(X,Y,color,levels=np.linspace(0,100,101),cmap='jet')
plt.axis('square')
plt.show()
```
